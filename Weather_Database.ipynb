```python
#importing dependencies
import pandas as pd
import numpy as np
import requests
from citipy import citipy
from config import weather_api_key
```

```python
#generating random numbers
lat = np.random.uniform(-90.00, 90.00, size=1500)
lng = np.random.uniform(-180.00, 180.00, size=1500)

lat_lngs = zip(lat, lng)
lat_lngs
```

```python
#generating list of cities with citipy
cities = []

for pair in lat_lngs:
    city = citipy.nearest_city(pair[0], pair[1]).city_name
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
test = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key+ "&q=" + "Mayumba"

print(test)

#datetime dependency
from datetime import datetime
```

```python
#openweather API call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

test = cities[:10]
# Create counters.
record_count = 1
set_count = 1

city_data = []
#populate list
for i, city in enumerate(cities):

    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    
    #endpoint url
    city_url = url + "&q=" + city
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    record_count += 1
    try:
        city_weather = requests.get(city_url).json()
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        weather_des = city_weather["weather"][0]['description']
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        try: 
            city_rain = city_weather["rain"]["3h"]
        except:
            city_rain = "0.0"
        try:
            city_snow = city_weather["snow"]["3h"]
        except:
            city_snow = "0.0"

        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Current Description": weather_des,
                          "Rain inches (last 3 hrs)": city_rain,
                          "Snow inches (last 3 hrs)": city_snow,
                          "Date": city_date})

    except:
        print("City not found. Skipping...")
        pass
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
len(city_data)
```

```python
#make city dataframe
city_data_df = pd.DataFrame(city_data)
new_column_order = ["City", "Country", "Date", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description", "Rain inches (last 3 hrs)", "Snow inches (last 3 hrs)"]
city_data_df = city_data_df[new_column_order]

#changing rain and snow dtypes to float64
city_data_df = city_data_df.astype({"Snow inches (last 3 hrs)": 'float'})
city_data_df = city_data_df.astype({"Rain inches (last 3 hrs)": 'float'})
city_data_df.dtypes

#save
file_to_save = "weather_data/weatherpy_challenge.csv"

city_data_df.to_csv(file_to_save, index_label="City_ID")
```
